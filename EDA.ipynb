```python
# fiiling the null values with median 
df["LB"]= df["LB"].fillna(df["LB"].median())
df["AC"]= df["AC"].fillna(df["AC"].median())
df["DS"]= df["DS"].fillna(df["DS"].median())
df["DP"]= df["DP"].fillna(df["DP"].median())
df["MLTV"]= df["MLTV"].fillna(df["MLTV"].median())
df["Width"]= df["Width"].fillna(df["Width"].median())
df["Tendency"]= df["Tendency"].fillna(df["Tendency"].median())
df["NSP"]= df["NSP"].fillna(df["NSP"].median())
```

```python
df.isna().sum()
```

```python
df.drop(columns=["DS"],inplace = True ) #dropping the uncessary column
df.drop(columns=["DP"],inplace =True)# droping the uncessary column
```

### 3)Removing the OutLiers

```python
sns.boxplot(df) # checking for outliers using boxplot
```

```python
def outlier_detection(df,columns):# function that removes an outlier
    for col in columns(df,columns):
        
        
        Q1 =df[col].quantile(0.25)
        Q3 =df[col].quantile(0.75)
        iqr = Q3-Q1
        lower,upper = Q1-1.5*iqr,Q3+1.5*IQR
        df[col] =np.clip(df[col],lower,upper)
    return df
    
```

```python
df = outlier_detection(df,['LB', 'AC', 'FM', 'UC', 'DL', 'ASTV', 'MSTV', 'ALTV',
       'MLTV', 'Width', 'Tendency', 'NSP']) # calling the function
```

```python
df.describe() #describe for checking the outliers
```

```python
sns.boxplot(df) # outliers are removed
```

```python
df.describe()
```
