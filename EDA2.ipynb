# Encoding

```python
from sklearn.preprocessing import LabelEncoder
# Identify categorical columns with less than 5 categories
categorical_cols_lt5 = [col for col in data1.select_dtypes(include=['object']).columns if data1[col].nunique() < 5]
# Apply One-Hot Encoding
data1 = pd.get_dummies(data1, columns=categorical_cols_lt5, drop_first=True)

# Identify categorical columns with more than 5 categories in the updated DataFrame
categorical_cols_gt5 = [col for col in data1.select_dtypes(include=['object']).columns if data1[col].nunique() >= 5]

# Initialize the LabelEncoder
label_encoders = {}

# Apply Label Encoding to these columns
for col in categorical_cols_gt5:
    le = LabelEncoder()
    # Fit and transform the column data
    data1[col] = le.fit_transform(data1[col])
    # Save the encoder for inverse transformation if needed
    label_encoders[col] = le
data1.head()
```

# Feature Engineering

```python
data1['age_income_ratio'] = data1['age'] / data1['fnlwgt'] 
```

```python
# Handle division by zero
data1['capital_gain_loss_ratio'] = np.where(data1['capital_loss'] == 0,
                                            data1['capital_gain'], data1['capital_gain'] / data1['capital_loss'])

data1.head()
```

```python
### transformation skewed numerical feature and justify your choice.
from scipy.stats import skew
# Calculate skewness 
numerical_cols = data1.select_dtypes(include=['int64', 'float64'])
# Calculate skewness for numerical columns
skewness = numerical_cols.apply(lambda x: skew(x)) 
print(skewness)
```

```python
# Apply log transformation
data1['capital_gain'] = np.log1p(data1['capital_gain']) # high skewed
data1['capital_loss'] = np.log1p(data1['capital_loss'])  # moderate but skewed
data1['fnlwgt'] = np.log1p(data['fnlwgt'])      # low skew
data1['age_income_ratio'] = np.log1p(data1['age_income_ratio'])   # high skew 
data1['capital_gain_loss_ratio'] = np.log1p(data1['capital_gain_loss_ratio'])   # high skewed
data1.head()
```

```python
# Handle NaN values if any
# For simplicity, you might replace NaN with a specific value, e.g., 0 or mean
data1.fillna(0, inplace=True)
```

# Feature Selection

```python
# Use the Isolation Forest algorithm to identify and remove outliers.

from sklearn.ensemble import IsolationForest
iso_forest = IsolationForest(contamination=0.1)  
outliers = iso_forest.fit_predict(data1)

# Convert to DataFrame
outliers_df = pd.DataFrame({'Outlier': outliers})
data_with_outliers = pd.concat([data1, outliers_df], axis=1)

# Display the data with outliers marked
print(data_with_outliers)
```

```python
# Filter out the outliers (rows where 'Outlier' == -1)
data_clean = data_with_outliers[data_with_outliers['Outlier'] != -1].drop(columns=['Outlier'])
data_clean.head()
```
