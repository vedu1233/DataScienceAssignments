```python
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
sns.boxplot(df[numerical_col[0]],ax=axs[0,0])
sns.boxplot(df[numerical_col[1]],ax=axs[0,1])
sns.boxplot(df[numerical_col[2]],ax=axs[0,2])
sns.boxplot(df[numerical_col[3]],ax=axs[1,0])
sns.boxplot(df[numerical_col[4]],ax=axs[1,1])
sns.boxplot(df[numerical_col[5]],ax=axs[1,2])
```

```python
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    median = df[col].median()
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    

    

    # Print the results
    print(f"Column: {col}")
    print(f"  Q1: {Q1}")
    print(f"  Q3: {Q3}")
    print(f"  Median: {median}")
    print(f"  IQR: {IQR}")
    print(f"  Outliers: {df[col][(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))].count()}")
    print()
    print()  
```

### Standardization of Numerical Variables

```python
from sklearn.preprocessing import StandardScaler
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df)
plt.title('Original Data Distribution')

# Standardize numerical columns
scaler = StandardScaler()
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# standardized distributions
plt.subplot(1, 2, 2)
sns.boxplot(data=df)
plt.title('Standardized Data Distribution')

# Display the plots
plt.tight_layout()
plt.show()

# Show the first few rows of the original and standardized data
print("Original Data:\n", df.head(5))
print("Standardized Data:\n", df.head(5))
```
